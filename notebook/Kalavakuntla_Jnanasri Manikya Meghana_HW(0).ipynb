```python
data_frame_dict=pd.read_csv('../data/Salaries.csv')
data_frame_dict
```

```python
dict_1=data_frame_dict.to_dict()
```

```python
key_names=data_frame_dict.columns
key_names
```

```python
dictionary = {};
for key in key_names:
    dictionary[key]=[]
dictionary
```

```python
for key in key_names:
    for index, row_data in data_frame_dict.iterrows():
        value_array =dictionary[key]
        value_array.append(row_data[key])
        dictionary[key]=value_array
```

```python
dictionary
```

```python
new_df=pd.DataFrame(dictionary)
```
